```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-15,15)
f1 = np.sin(x)/x
f2 = 2 * np.sin(x)/x
f3 = 3 * np.sin(x)/x
```

```python
plt.plot(x,f1,color='blue',linestyle='--', label='sin(x)/x')
plt.plot(x,f2,color='green',linestyle='-.', label='2sin(x)/x')
plt.plot(x,f3,color='red',marker='o', label='3sin(x)/x')
plt.legend()
```
