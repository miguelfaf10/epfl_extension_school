__Simple linear regressions__

```python
import pandas as pd

# Load the dataset
data_df = pd.read_csv("c3_marketing-campaign.csv")

# Extract x/y values
x = data_df.tv.values
y = data_df.sales.values

print("x shape:", x.shape)
print("y shape:", y.shape)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# Plot data
plt.scatter(x, y)
plt.xlabel("tv budget in 1000$")
plt.ylabel("sales in thousands")
plt.show()
```

__Polyfit function__

import numpy as np

# Fit a linear regression

```python
import numpy as np

# Fit a linear regression
coefs = np.polyfit(x, y, deg=1)
print("Coefficients:", coefs)
```

```python
# Generate a hundred data points
x_values = np.linspace(x.min(), x.max(), num=100)

# Compute predictions
y_values = np.polyval(coefs, x_values)
```

```python
# Plot predictions
plt.scatter(x, y)
plt.plot(x_values, y_values, c="C3", label="linear regression")
plt.xlabel("tv budget in 1000$")
plt.ylabel("sales in thousands")
plt.legend()
plt.show()
```
